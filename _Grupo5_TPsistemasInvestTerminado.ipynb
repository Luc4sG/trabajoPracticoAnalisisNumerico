## 1. NumPy: Operaciones con Matrices
NumPy es una biblioteca fundamental en Python para el c√°lculo num√©rico y el procesamiento de grandes vol√∫menes de datos. Ofrece un conjunto de funciones y estructuras de datos optimizadas para operaciones matem√°ticas, principalmente a trav√©s de su estructura central: el array de NumPy (similar a las matrices en matem√°ticas). Los arrays de NumPy son altamente eficientes en t√©rminos de memoria y rendimiento, lo que permite realizar c√°lculos sobre grandes conjuntos de datos de forma r√°pida y con menor uso de recursos.

Para multiplicar matrices utilizamos la funcion numpy.dot() o el operador @.

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = np.dot(A, B)  # o C = A @ B
print(C)  # [[19 22] [43 50]]
```

### C√°lculo de la Inversa de una Matriz: Usamos numpy.linalg.inv() para obtener la inversa.


```python
A = np.array([[1, 2], [3, 4]])
A_inv = np.linalg.inv(A)
print(A_inv)  # [[-2.   1. ] [ 1.5 -0.5]]
```

### C√°lculo de Valores Propios y Vectores Propios: Utilizamos numpy.linalg.eig().

```python
A = np.array([[4, -2], [1, 1]])
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Valores propios:", eigenvalues)
print("Vectores propios:", eigenvectors)
```

## 2. Resolver un Sistema de Ecuaciones Lineales
Para resolver un sistema de ecuaciones lineales usando matrices en NumPy, primero debemos expresar el sistema en forma matricial. Un sistema de ecuaciones como: 

2x + 3y = 8   
 x - 2y = -3

se puede escribir en t√©rminos de matrices como:
A * x = B

donde:
A es la matriz de coeficientes [[2, 3], [1, -2]]
B es el vector de resultados [8, -3]
Para resolver este sistema en NumPy, usamos numpy.linalg.solve(A, B), que devuelve la soluci√≥n para 'X' y 'Y' sin necesidad de calcular la inversa de A.

```python
import numpy as np

# Definimos la matriz de coeficientes A y el vector de resultados B
A = np.array([[2, 3], [1, -2]])
B = np.array([8, -3])

# Resolvemos el sistema de ecuaciones
solution = np.linalg.solve(A, B)

# Mostramos la soluci√≥n para x y y
print("Soluci√≥n:", solution)
```

## 3. SymPy y C√°lculo Diferencial
SymPy es una biblioteca de Python dedicada al c√°lculo simb√≥lico, que permite realizar operaciones matem√°ticas de forma anal√≠tica en lugar de num√©rica. A diferencia de otras bibliotecas como NumPy o SciPy, que se enfocan en c√°lculos num√©ricos, SymPy manipula expresiones matem√°ticas simb√≥licamente, tal como lo har√≠a un matem√°tico. Esto permite simplificar expresiones, derivar funciones, integrar y resolver ecuaciones diferenciales de manera exacta y sin aproximaciones.

**En el √°mbito del c√°lculo diferencial, SymPy es especialmente √∫til en √°reas como:**

**Derivaci√≥n y Diferenciaci√≥n Autom√°tica:** Permite calcular derivadas de expresiones simb√≥licas, simplificar resultados y manipular expresiones derivadas para an√°lisis adicionales.
Integraci√≥n y Resoluci√≥n de Integrales: SymPy permite resolver integrales indefinidas y definidas, simplificar expresiones integrales y calcular primitivas simb√≥licamente.
Ecuaciones Diferenciales: SymPy es una herramienta poderosa para la resoluci√≥n de ecuaciones diferenciales ordinarias (EDO) y parciales (EDP) de forma simb√≥lica. A trav√©s de sus funciones, permite encontrar soluciones exactas cuando existen y explorar soluciones generales o particulares.
SymPy en la Resoluci√≥n de Ecuaciones Diferenciales
La biblioteca SymPy incluye funciones espec√≠ficas para resolver ecuaciones diferenciales, como dsolve(), que permite resolver EDOs, y herramientas para trabajar con EDPs. Al utilizar SymPy en lugar de m√©todos puramente num√©ricos, obtenemos ventajas significativas, como:

**Soluciones Exactas:** SymPy permite obtener soluciones exactas (simb√≥licas), mientras que los m√©todos num√©ricos, como los m√©todos de Euler o Runge-Kutta, solo proporcionan aproximaciones.
**An√°lisis Completo de la Soluci√≥n:** Las soluciones simb√≥licas facilitan la comprensi√≥n de la estructura matem√°tica de la respuesta, permitiendo simplificar, analizar y derivar nuevas relaciones.
**Manipulaci√≥n de Expresiones:** Con una soluci√≥n simb√≥lica, es posible manipular la expresi√≥n final para obtener derivadas, integrales adicionales u otras transformaciones sin p√©rdida de precisi√≥n.
**Ejemplo de SymPy Resolviendo una Ecuaci√≥n Diferencial**
A continuaci√≥n, se muestra un ejemplo de c√≥mo SymPy puede resolver simb√≥licamente una ecuaci√≥n diferencial ordinaria.

```python
import sympy as sp

# Definimos la variable independiente y la funci√≥n dependiente
t = sp.symbols('t')
y = sp.Function('y')(t)

# Definimos la ecuaci√≥n diferencial: y' + y = e^t
equation = sp.Eq(y.diff(t) + y, sp.exp(t))

# Resolvemos la ecuaci√≥n diferencial
solution = sp.dsolve(equation)

# Mostramos la soluci√≥n
solution
```

Este c√≥digo define y resuelve la ecuaci√≥n diferencial ordinaria **ùë¶‚Ä≤+ùë¶=ùëíùë°** proporcionando la soluci√≥n simb√≥lica exacta. La capacidad de obtener esta soluci√≥n anal√≠tica es una ventaja de SymPy frente a m√©todos que solo pueden aproximar soluciones num√©ricamente.

## 4. SciPy: Resoluci√≥n de Ecuaciones Diferenciales

SciPy es una biblioteca de Python construida sobre NumPy, dise√±ada para resolver problemas matem√°ticos, cient√≠ficos y de ingenier√≠a. Mientras que NumPy proporciona estructuras de datos r√°pidas y eficientes, SciPy ampl√≠a estas capacidades con una amplia gama de herramientas y algoritmos especializados, como integraci√≥n, optimizaci√≥n, √°lgebra lineal avanzada y estad√≠sticas. SciPy ofrece herramientas para resolver ecuaciones diferenciales de forma num√©rica, enfoc√°ndose en m√©todos eficientes y precisos. Dentro de SciPy, las funciones m√°s utilizadas para resolver ecuaciones diferenciales son odeint y solve_ivp. Ambas se encuentran en el subm√≥dulo scipy.integrate y son especialmente √∫tiles para resolver ecuaciones diferenciales ordinarias (EDOs). A continuaci√≥n, se describen sus principales caracter√≠sticas y diferencias.

**Funci√≥n odeint**: odeint es una funci√≥n ampliamente utilizada para resolver sistemas de ecuaciones diferenciales de primer orden.
**Uso:** Requiere definir la funci√≥n que describe el sistema de ecuaciones diferenciales y acepta par√°metros adicionales que se pueden pasar a dicha funci√≥n.
**Ventaja:** odeint es sencilla y r√°pida para resolver problemas de ecuaciones diferenciales relativamente simples y bien comportados.
**Limitaci√≥n:** Dado que est√° basada en un c√≥digo antiguo, puede ser menos flexible y menos compatible con sistemas complejos o configuraciones avanzadas de integraci√≥n.

```python
from scipy.integrate import odeint

# Definimos la funci√≥n del sistema de ecuaciones
def modelo(y, t):
    dydt = -2 * y + 3
    return dydt

# Condici√≥n inicial y valores de tiempo
y0 = 5
t = np.linspace(0, 10, 100)

# Resoluci√≥n del sistema con odeint
sol = odeint(modelo, y0, t)
```

**Funci√≥n solve_ivp**
**Descripci√≥n:** solve_ivp (solve initial value problem) es una funci√≥n m√°s moderna y flexible para resolver EDOs, que incluye varios m√©todos de integraci√≥n (como Runge-Kutta de orden 5, RK45, y Radau).
**Uso:** Similar a odeint, requiere definir la funci√≥n del sistema, pero permite especificar el m√©todo de integraci√≥n y el intervalo de soluci√≥n.
**Ventaja:** solve_ivp es m√°s flexible y compatible con problemas m√°s complejos, adem√°s de que admite varias opciones de m√©todos y configuraciones avanzadas.
**Limitaci√≥n:** Aunque ofrece m√∫ltiples m√©todos, puede requerir un ajuste cuidadoso de par√°metros para obtener un rendimiento √≥ptimo.

```python
from scipy.integrate import solve_ivp

# Definimos la funci√≥n del sistema de ecuaciones
def modelo(t, y):
    dydt = -2 * y + 3
    return dydt

# Condici√≥n inicial y resoluci√≥n con solve_ivp
t_span = (0, 10)   # Intervalo de tiempo
y0 = [5]           # Condici√≥n inicial
sol = solve_ivp(modelo, t_span, y0, method='RK45')
```

**¬øCu√°ndo usar odeint o solve_ivp?**
odeint es una buena opci√≥n para problemas est√°ndar o simples, donde se necesita un enfoque r√°pido y sin configuraciones avanzadas.
solve_ivp es preferible en problemas complejos, donde se necesite control sobre el m√©todo de integraci√≥n, o cuando se busque una soluci√≥n en un rango espec√≠fico con pasos de tiempo adaptativos.

## 5. Graficar Campos Vectoriales con Matplotlib

**Matplotlib** es una biblioteca de Python para la visualizaci√≥n de datos, permitiendo la creaci√≥n de gr√°ficos de l√≠neas, histogramas, gr√°ficos de dispersi√≥n y muchos otros tipos de visualizaci√≥n. Es ampliamente usada en ciencia de datos, an√°lisis y visualizaci√≥n cient√≠fica por su flexibilidad y facilidad de uso.

**Graficar Campos Vectoriales con Matplotlib**
Para graficar campos vectoriales en Matplotlib, se usa la funci√≥n quiver, que permite representar un campo de vectores en un plano 2D. Un campo vectorial es una representaci√≥n visual de vectores en diferentes puntos, que indican direcci√≥n y magnitud en una regi√≥n del espacio.

**Funci√≥n quiver**
La funci√≥n quiver de Matplotlib se utiliza para trazar flechas en cada punto de un campo, representando vectores en coordenadas X e Y. La flecha apunta en la direcci√≥n del vector, y su tama√±o representa la magnitud del vector.

**Sintaxis B√°sica**

import matplotlib.pyplot as plt
import numpy as np

plt.quiver(X, Y, U, V, kwargs)

**Par√°metros Principales**
X, Y: Arrays de coordenadas donde se colocar√°n las colas de los vectores. Estos son arreglos de las coordenadas en el eje X y Y.

U, V: Arrays de componentes vectoriales en las direcciones X e Y, respectivamente. Es decir, U y V definen la direcci√≥n y magnitud de cada vector en el campo.

**Argumentos Opcionales (kwargs)**

scale: Escala para los vectores. Permite ajustar la longitud de las flechas; si se omite, Matplotlib ajusta la escala autom√°ticamente.

scale_units: Unidades en las que se basa la escala (por ejemplo, "width" o "height").

angles: Direcci√≥n de los vectores (por ejemplo, "xy" hace que las flechas apunten en la direcci√≥n (U, V)).

color: Color de las flechas.

width: Grosor de las flechas.

```python
import matplotlib.pyplot as plt
import numpy as np

# Definimos una cuadr√≠cula de puntos en el espacio
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Definimos los componentes del campo vectorial
U = -Y  # Componente en X
V = X   # Componente en Y

# Creamos el gr√°fico del campo vectorial
plt.quiver(X, Y, U, V, color='blue', scale=20)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Campo Vectorial usando quiver')
plt.grid()
plt.show()
```

# Parte 2: Aplicaciones Num√©ricas y Simb√≥licas

A continuacion se resolveremos la siguiente ecuacion diferencial: y'' + 9y = e^t, y(0) = 0, y'(0) = 0

En el enfoque simb√≥lico, utilizaremos la funci√≥n dsolve de **SymPy** para obtener una soluci√≥n exacta de la ecuaci√≥n diferencial. Este enfoque es ideal cuando queremos una soluci√≥n exacta y una comprensi√≥n clara de la forma matem√°tica de la respuesta.

```python
import sympy as sp

# Definimos las variables y la funci√≥n
t = sp.symbols('t')
y = sp.Function('y')(t)

# Definimos la ecuaci√≥n diferencial
eq = sp.Eq(y.diff(t, t) + 9 * y, sp.exp(t))

# Condiciones iniciales
ics = {y.subs(t, 0): 0, y.diff(t).subs(t, 0): 0}

# Resoluci√≥n de la ecuaci√≥n diferencial
solution_sym = sp.dsolve(eq, y, ics=ics)

# Mostrar la soluci√≥n simb√≥lica
solution_sym
```

Para resolver esta ecuaci√≥n diferencial de forma num√©rica, usaremos la funci√≥n solve_ivp de **SciPy**. Aqu√≠ debemos expresar la ecuaci√≥n diferencial como un sistema de primer orden para que sea compatible con los m√©todos num√©ricos de SciPy.

```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

# Definimos el sistema de primer orden
def system(t, Y):
    y, dy = Y
    d2y = np.exp(t) - 9 * y
    return [dy, d2y]

# Condiciones iniciales
y0 = [0, 0]  # y(0) = 0, y'(0) = 0

# Intervalo de tiempo
t_span = (0, 5)
t_eval = np.linspace(0, 5, 100)

# Resoluci√≥n num√©rica del sistema
solution_num = solve_ivp(system, t_span, y0, t_eval=t_eval)

# Graficar la soluci√≥n num√©rica
plt.plot(solution_num.t, solution_num.y[0], label="y(t) - Soluci√≥n Num√©rica")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Soluci√≥n Num√©rica con SciPy")
plt.legend()
plt.show()
```

**Comparaci√≥n de los Enfoques**

**SymPy (Enfoque Simb√≥lico):**

Ventaja: Proporciona una soluci√≥n exacta en t√©rminos matem√°ticos, √∫til para an√°lisis te√≥ricos y cuando queremos una representaci√≥n algebraica de la respuesta.
Desventaja: Puede ser ineficiente o imposible de resolver si la ecuaci√≥n es demasiado compleja o no tiene una soluci√≥n simb√≥lica.

**SciPy (Enfoque Num√©rico):**

Ventaja: Es muy r√°pido y eficiente para ecuaciones diferenciales complejas o para sistemas de ecuaciones que no tienen soluci√≥n exacta.
Desventaja: La soluci√≥n es aproximada y depende del intervalo de tiempo y el m√©todo de integraci√≥n, por lo que no es ideal cuando se necesita una soluci√≥n exacta.

**Conclusiones**
Se utiliza SymPy cuando se necesita una soluci√≥n exacta y la ecuaci√≥n es manejable de forma simb√≥lica. En cambio, se puede optar por SciPy en casos donde la ecuaci√≥n es compleja, el sistema tiene muchas variables, o se necesita una soluci√≥n r√°pida y aproximada en un intervalo espec√≠fico de tiempo.

## 2. Transformada de Laplace en SymPy
La biblioteca SymPy permite calcular la Transformada de Laplace y su inversa de forma simb√≥lica, lo cual es √∫til para trabajar con ecuaciones diferenciales y sistemas din√°micos.

#### Uso de las funciones laplace_transform e inverse_laplace_transform en SymPy laplace_transform ####

#### Laplace_Transform ####
La funci√≥n **laplace_transform** en SymPy calcula la Transformada de Laplace de una funci√≥n en el dominio del tiempo.

**Sintaxis:**
**laplace_transform(f, t, s)** 

**Par√°metros:**

f: la funci√≥n a transformar, expresada en t√©rminos de 'ùë°' (tiempo).

t: la variable del dominio del tiempo (generalmente, ùë°).

s: la variable del dominio de la frecuencia (generalmente, ùë†).

Retorno: Devuelve una tupla con tres elementos: la transformada de Laplace de la funci√≥n, la regi√≥n de convergencia y una condici√≥n para la variable ùë†.

En el siguiente ejemplo, la funci√≥n sp.laplace_transform() calcula la Transformada de Laplace de **ùëí^‚àí2ùë°**, transformando la funci√≥n al dominio ùë†.

```python
import sympy as sp

# Definimos la variable y la funci√≥n
t, s = sp.symbols('t s')
f = sp.exp(-2 * t)

# Calculamos la Transformada de Laplace
laplace_f = sp.laplace_transform(f, t, s)
laplace_f
```

#### inverse_laplace_transform ####

La funci√≥n inverse_laplace_transform calcula la Transformada de Laplace inversa, devolviendo la funci√≥n en el dominio del tiempo.

**Sintaxis:
inverse_laplace_transform(F, s, t)**

Par√°metros:

F: la funci√≥n en el dominio de la frecuencia (en t√©rminos de ùë†) a transformar de vuelta al dominio del tiempo.
s: la variable del dominio de la frecuencia.
t: la variable del dominio del tiempo.
Retorno: Devuelve la funci√≥n en el dominio del tiempo correspondiente a la Transformada de Laplace inversa.

Ejemplo:

```python
# Definimos la funci√≥n en el dominio de la frecuencia
F = 1 / (s + 2)

# Calculamos la Transformada de Laplace inversa
inverse_laplace_f = sp.inverse_laplace_transform(F, s, t)
inverse_laplace_f
```

#### Conclusiones ####

**La Transformada de Laplace es especialmente √∫til en √°reas como:**

**Resoluci√≥n de Ecuaciones Diferenciales:** Es una herramienta clave para resolver ecuaciones diferenciales lineales con condiciones iniciales. Al transformar la ecuaci√≥n diferencial en una ecuaci√≥n algebraica en el dominio de ùë†, el problema se simplifica considerablemente.

**An√°lisis de Sistemas de Control:** En ingenier√≠a de control, la Transformada de Laplace permite analizar sistemas de control lineales. Facilita el dise√±o y estudio de la estabilidad y respuesta en el tiempo de sistemas de control.

**Procesamiento de Se√±ales:** Permite analizar y manipular se√±ales en el dominio de la frecuencia, siendo fundamental en telecomunicaciones y electr√≥nica para estudiar el comportamiento de filtros y sistemas de transmisi√≥n.

**F√≠sica e Ingenier√≠a El√©ctrica:** Se usa en circuitos el√©ctricos y sistemas din√°micos para modelar y resolver problemas que involucran elementos como resistencias, capacitores e inductores.

La capacidad de SymPy para realizar tanto la Transformada de Laplace como su inversa de manera simb√≥lica permite un an√°lisis te√≥rico y exacto de estos sistemas, lo cual es especialmente √∫til en contextos acad√©micos, investigaci√≥n y en la primera fase de dise√±o de sistemas antes de realizar simulaciones num√©ricas m√°s complejas.

## 3. Construcci√≥n y Manipulaci√≥n de Matrices Simb√≥licas en SymPy
En SymPy, puedes crear matrices simb√≥licas usando la clase Matrix. Esto es √∫til cuando quieres trabajar con variables en lugar de n√∫meros espec√≠ficos.

Para construir matrices simb√≥licas en SymPy, se utilizan las clases Matrix o MutableMatrix, que permiten definir matrices con entradas simb√≥licas. Primero, es necesario definir las variables simb√≥licas y luego crear la matriz especificando sus elementos.

Ejemplo B√°sico:
Aqu√≠ creamos una matriz simb√≥lica ùê¥ con elementos que son variables simb√≥licas: 

```python
import sympy as sp

# Definimos variables simb√≥licas
a, b, c, d = sp.symbols('a b c d')

# Creamos una matriz simbolica 'A' con entradas a, b, c y d.
A = sp.Matrix([[a, b], [c, d]])

# Mostramos la matriz
A
```

#### Ejemplo de Matrices Simb√≥licas en Sistemas de Ecuaciones ####
Las matrices simb√≥licas son √∫tiles para resolver sistemas de ecuaciones lineales de manera simb√≥lica. Por ejemplo, para el sistema de ecuaciones:

a*x + b*y = 1
c*x + d*y = 0

Podemos representar el sistema como una ecuaci√≥n matricial A * X = B, donde:

A = Matrix([[a, b], [c, d]])
X = Matrix([x, y])
B = Matrix([1, 0])

Podemos resolver el sistema con SymPy de esta manera:

```python
# Definimos las variables simb√≥licas para las inc√≥gnitas
x, y = sp.symbols('x y')

# Definimos la matriz de coeficientes A y el vector de resultados B
A = sp.Matrix([[a, b], [c, d]])
B = sp.Matrix([1, 0])

# Calculamos la soluci√≥n del sistema A * X = B
solution = A.solve_least_squares(B)

# Mostramos la soluci√≥n para x y y
solution
```

## 4. C√°lculo del Determinante y la Inversa de una Matriz Simb√≥lica
En SymPy, podemos calcular el determinante y la inversa de una matriz simb√≥lica usando los m√©todos .det() y .inv().

Ejemplo practico:

```python
# C√°lculo del determinante
det_A = A.det()
print("Determinante de A:", det_A)

# C√°lculo de la inversa
inv_A = A.inv()
print("Inversa de A:", inv_A)
```

## 5. Ventajas de SymPy sobre NumPy para Matrices Simb√≥licas
SymPy y NumPy tienen enfoques diferentes al trabajar con matrices:

Aunque con ambas se puede trabajar con matrices, ofrecen capacidades y enfoques diferentes cuando se trata de matrices simb√≥licas. 

**C√°lculo simb√≥lico:** SymPy est√° dise√±ada para el c√°lculo simb√≥lico, permitiendo trabajar con variables y realizar manipulaciones algebraicas en lugar de valores num√©ricos concretos. Esto es fundamental en problemas donde las entradas de la matriz son expresiones simb√≥licas (como variables, constantes algebraicas, o funciones). NumPy, en cambio, est√° enfocado en c√°lculos num√©ricos y no tiene soporte nativo para operaciones simb√≥licas.

**Manipulaci√≥n y simplificaci√≥n algebraica:** SymPy permite simplificar expresiones, expandir polinomios, factorizar, y reordenar t√©rminos en las entradas de la matriz. Esto es especialmente √∫til para obtener soluciones en su forma m√°s simple o para reformular ecuaciones de sistemas lineales. NumPy no tiene capacidades para manipulaci√≥n algebraica.

**Operaciones avanzadas con matrices simb√≥licas:** Con SymPy, se pueden calcular derivadas, integrales y l√≠mites directamente sobre las entradas simb√≥licas de una matriz. Esto es √∫til para problemas de optimizaci√≥n simb√≥lica, an√°lisis de sensibilidad, y en contextos matem√°ticos avanzados, como √°lgebra y c√°lculo multivariable. NumPy, al ser num√©rica, no permite estas operaciones sin la ayuda de otras librer√≠as.

**Soluci√≥n de sistemas lineales simb√≥licos:** SymPy permite resolver sistemas de ecuaciones lineales donde tanto los coeficientes como las soluciones son simb√≥licas. NumPy puede resolver sistemas lineales, pero solo con coeficientes num√©ricos.

**Resultados exactos:** SymPy trabaja con aritm√©tica exacta, proporcionando resultados en fracciones y expresiones algebraicas exactas en lugar de valores aproximados, lo cual es ideal en contextos donde la precisi√≥n simb√≥lica es importante. NumPy, al ser num√©rica, siempre devuelve valores aproximados (en punto flotante), que pueden ser propensos a errores de redondeo.

**Conclusiones**

## 1. C√°lculo de Derivadas Simb√≥licas en SymPy
La biblioteca SymPy permite calcular derivadas simb√≥licas usando la funci√≥n diff, tanto para derivadas simples como parciales.

#### Derivadas Simples con diff ####
Para calcular una derivada simple, define la funci√≥n en t√©rminos de una variable y aplica diff con respecto a esa variable.

Ejemplo: Derivada Simple

Calcula la derivada de f(x) = x^3 + 2x^2 + x con respecto a x:

```python
import sympy as sp

# Definimos la variable y la funci√≥n
x = sp.symbols('x')
f = x**3 + 2*x**2 + x

# Calculamos la derivada de f con respecto a x
f_prime = sp.diff(f, x)
f_prime
```

#### Derivadas Parciales con diff ####
Para derivadas parciales, define la funci√≥n en t√©rminos de varias variables y aplica diff con respecto a la variable que deseas derivar.

Ejemplo: Derivada Parcial

Calcula la derivada parcial de g(x, y) = x^2 * y + y^2 * x con respecto a x y luego con respecto a y:

```python
# Definimos las variables y la funci√≥n
x, y = sp.symbols('x y')
g = x**2 * y + y**2 * x

# Derivada parcial de g con respecto a x
g_x = sp.diff(g, x)

# Derivada parcial de g con respecto a y
g_y = sp.diff(g, y)

g_x, g_y
```

## 2. C√°lculo de L√≠mites Simb√≥licos en SymPy
En SymPy, se pueden calcular l√≠mites simb√≥licos de funciones usando la funci√≥n limit. Esta funci√≥n es muy √∫til para calcular l√≠mites de manera exacta sin realizar aproximaciones num√©ricas.

**Uso de la Funci√≥n limit**
La funci√≥n limit toma tres argumentos principales:

func: la funci√≥n de la que deseas calcular el l√≠mite.

var: la variable que tiende a un valor.

value: el valor al que se acerca la variable.

**Sintaxis: limit(func, var, value)**

SymPy tambi√©n permite calcular l√≠mites laterales usando los argumentos opcionales dir, donde:

dir='+' calcula el l√≠mite cuando la variable tiende al valor desde la derecha (l√≠mite por la derecha).

dir='-' calcula el l√≠mite cuando la variable tiende al valor desde la izquierda (l√≠mite por la izquierda).

**Ejemplo: C√°lculo de un L√≠mite**
Calculemos el l√≠mite de la funci√≥n f(x) = (x**2 - 1) / (x - 1) cuando x tiende a 1.

```python
import sympy as sp

# Definimos la variable y la funci√≥n
x = sp.symbols('x')
f = (x**2 - 1) / (x - 1)

# Calculamos el l√≠mite cuando x tiende a 1
limite = sp.limit(f, x, 1)
limite
```

**Ejemplo de L√≠mite Lateral con argumento opcional dir**

Calculemos el l√≠mite de 1 / x cuando x tiende a 0 desde la derecha (dir='+').

```python
# Calculamos el l√≠mite cuando x tiende a 0 por la derecha
limite_derecha = sp.limit(1 / x, x, 0, dir='+')
limite_derecha
```

## 4. Importancia enfoque simbolico ##

Un caso en el que el c√°lculo simb√≥lico de l√≠mites es m√°s √∫til que la evaluaci√≥n num√©rica es cuando se enfrenta a una indeterminaci√≥n, como 0/0 o ‚àû/‚àû. Estas indeterminaciones aparecen con frecuencia en matem√°ticas y f√≠sica, especialmente al analizar comportamientos de funciones cerca de puntos cr√≠ticos o en el infinito.

#### Ejemplo: Funciones con Indeterminaci√≥n ####
Consideremos la funci√≥n f(x) = (sin(x) / x) cuando x tiende a 0. Si intentamos evaluar esta funci√≥n num√©ricamente en x = 0, obtenemos una indeterminaci√≥n 0/0. Un enfoque num√©rico (como sustituir un valor muy cercano a cero) no da una respuesta precisa y puede resultar en valores muy diferentes, lo cual no garantiza la exactitud de la soluci√≥n.

En este caso, el enfoque simb√≥lico es necesario porque:

**El C√°lculo Simb√≥lico Permite Simplificar Expresiones Algebraicas:** SymPy, al calcular simb√≥licamente el l√≠mite, puede reconocer y simplificar la expresi√≥n sin(x) / x usando identidades trigonom√©tricas. Al hacerlo, encuentra que el l√≠mite de sin(x) / x cuando x tiende a 0 es exactamente 1, una soluci√≥n que el c√°lculo num√©rico no puede encontrar con precisi√≥n debido a la indeterminaci√≥n.

**Generalizaci√≥n de Resultados:** El enfoque simb√≥lico da una expresi√≥n exacta del l√≠mite, que es √∫til en teor√≠a matem√°tica y en demostraciones. Por ejemplo, en c√°lculo, el l√≠mite de sin(x) / x en 0 es fundamental en series de Taylor y en an√°lisis de funciones peri√≥dicas. Obtener el l√≠mite simb√≥lico permite usar el resultado en contextos te√≥ricos sin depender de aproximaciones.

**Ejemplo con SymPy:**
El siguiente c√°lculo devuelve 1, que es el valor exacto del l√≠mite de la funci√≥n sin(x) / x en x = 0.

```python
import sympy as sp

# Definimos la variable y la funci√≥n
x = sp.symbols('x')
f = sp.sin(x) / x

# Calculamos el l√≠mite cuando x tiende a 0
limite = sp.limit(f, x, 0)
limite
```

#### Raz√≥n por la que el Enfoque Simb√≥lico es Necesario ####
El c√°lculo simb√≥lico es esencial en este tipo de problemas porque:

**Evita Errores de Aproximaci√≥n:** Los m√©todos num√©ricos pueden dar resultados inexactos o incluso divergentes al evaluar puntos cercanos a 0 en casos de indeterminaci√≥n.

**Garantiza la Exactitud en el An√°lisis Matem√°tico:** En situaciones donde el comportamiento de la funci√≥n en puntos cr√≠ticos afecta el resultado general (como en optimizaci√≥n o an√°lisis de l√≠mites en f√≠sica), el enfoque simb√≥lico asegura precisi√≥n y consistencia.

##  1. Ec Diferenciales con ScyPy utilizando solve_ivp ##

En SciPy, la funci√≥n **solve_ivp (solve initial value problem)** del m√≥dulo scipy.integrate es utilizada para resolver sistemas de ecuaciones diferenciales, incluidos aquellos que est√°n acoplados. solve_ivp facilita la resoluci√≥n de estos sistemas permitiendo obtener soluciones num√©ricas precisas en un intervalo de tiempo definido. 

**Pasos Clave para Resolver Sistemas Acoplados de Ecuaciones Diferenciales con solve_ivp**

Definir el Sistema de Ecuaciones:
Representamos el sistema de ecuaciones diferenciales acopladas como una funci√≥n que devuelve las derivadas de cada variable dependiente con respecto a la variable independiente, usualmente el tiempo t.

Especificar Condiciones Iniciales:
Definimos un vector con las condiciones iniciales de cada variable dependiente. Esto permite a solve_ivp comenzar la integraci√≥n desde un punto espec√≠fico en el tiempo.

Configurar el Intervalo de Integraci√≥n:
Definimos el rango de valores para la variable independiente (por ejemplo, tiempo) sobre el cual se resolver√°n las ecuaciones.

Elegir un M√©todo de Integraci√≥n (opcional):
solve_ivp permite seleccionar m√©todos espec√≠ficos de integraci√≥n (como RK45 o Radau). Por defecto, utiliza RK45, un m√©todo de Runge-Kutta de orden 5.

**Ejemplo:Consideremos el siguiente sistema de ecuaciones acopladas:**

dy1/dt = y2

dy2/dt = -y1

Este sistema describe un movimiento oscilatorio simple y puede resolverse con solve_ivp.


```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

# Definimos el sistema de ecuaciones como una funci√≥n
def sistema(t, Y):
    y1, y2 = Y  # Desempaquetamos las variables
    dy1_dt = y2
    dy2_dt = -y1
    return [dy1_dt, dy2_dt]

# Condiciones iniciales y configuraci√≥n del tiempo
condiciones_iniciales = [1, 0]  # y1(0) = 1, y2(0) = 0
t_span = (0, 10)  # Intervalo de tiempo
t_eval = np.linspace(0, 10, 100)  # Puntos de evaluaci√≥n en el tiempo

# Resolvemos el sistema
solucion = solve_ivp(sistema, t_span, condiciones_iniciales, t_eval=t_eval)

# Graficamos los resultados
plt.plot(solucion.t, solucion.y[0], label='y1(t)')
plt.plot(solucion.t, solucion.y[1], label='y2(t)')
plt.xlabel('Tiempo t')
plt.ylabel('Valores de y1 y y2')
plt.legend()
plt.title('Soluci√≥n del Sistema de Ecuaciones Diferenciales Acopladas')
plt.grid()
plt.show()
```

#### Explicaci√≥n de los Pasos Clave ####
Definici√≥n del Sistema: En el ejemplo, la funci√≥n sistema define las derivadas dy1_dt y dy2_dt en t√©rminos de y1 y y2. Estas derivadas representan el sistema acoplado de ecuaciones diferenciales.

Condiciones Iniciales: El vector [1, 0] representa y1(0) = 1 y y2(0) = 0, que son los valores iniciales de y1 y y2 en t = 0.

Intervalo y Evaluaci√≥n en Tiempo: t_span = (0, 10) indica que queremos resolver el sistema entre t = 0 y t = 10. t_eval define los puntos de tiempo donde se eval√∫an los valores de y1 y y2 para graficarlos.

Resoluci√≥n con solve_ivp: La funci√≥n solve_ivp integra el sistema y devuelve una soluci√≥n que se almacena en solucion. Las salidas solucion.t y solucion.y contienen los valores de t y las soluciones y1 y y2 en cada punto evaluado.
