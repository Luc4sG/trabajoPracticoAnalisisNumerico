```python
import numpy as np
import matplotlib.pyplot as plt

# Función para calcular el campo vectorial
def campo_vectorial(A, B, C, D, x, y):
    u_prime = A * x + B * y
    v_prime = C * x + D * y
    return u_prime, v_prime

# Parámetros para la cuadrícula
x = np.linspace(-10, 10, 20)
y = np.linspace(-10, 10, 20)
X, Y = np.meshgrid(x, y)

# Configuración de los casos
casos = {
    'Nodo Propio': {'A': -2, 'B': 0, 'C': 0, 'D': -3},
    'Foco Inestable': {'A': 1, 'B': -5, 'C': 5, 'D': 1},
    'Centro': {'A': 0, 'B': -1, 'C': 1, 'D': 0},
    'Punto de silla': {'A': 1, 'B': 1, 'C': 1, 'D': -1}
}

# Graficar los 4 sistemas
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
axes = axes.flatten()

for idx, (nombre, coef) in enumerate(casos.items()):
    A, B, C, D = coef['A'], coef['B'], coef['C'], coef['D']
    U, V = campo_vectorial(A, B, C, D, X, Y)
    
    # Crear el gráfico del campo vectorial
    axes[idx].quiver(X, Y, U, V, color='blue')
    axes[idx].set_title(f"Tipo: {nombre}")
    axes[idx].set_xlabel("u")
    axes[idx].set_ylabel("v")
    axes[idx].grid(True)
    axes[idx].set_aspect('equal')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Función para calcular el campo vectorial
def campo_vectorial(A, B, C, D, x, y):
    u_prime = A * x + B * y
    v_prime = C * x + D * y
    return u_prime, v_prime

# Parámetros para la cuadrícula con mayor resolución
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# Configuración de los casos
casos = {
    'Nodo Propio': {'A': -2, 'B': 0, 'C': 0, 'D': -3},
    'Foco Inestable': {'A': 1, 'B': -5, 'C': 5, 'D': 1},
    'Centro': {'A': 0, 'B': -1, 'C': 1, 'D': 0},
    'Punto de silla': {'A': 1, 'B': 1, 'C': 1, 'D': -1}
}

# Mejorar los gráficos
fig, axes = plt.subplots(2, 2, figsize=(14, 14), dpi=100)
axes = axes.flatten()

for idx, (nombre, coef) in enumerate(casos.items()):
    A, B, C, D = coef['A'], coef['B'], coef['C'], coef['D']
    U, V = campo_vectorial(A, B, C, D, X, Y)
    
    # Utilizar streamplot para un campo vectorial más suave
    axes[idx].streamplot(X, Y, U, V, color='teal', density=1.2, linewidth=1, arrowsize=1.5)
    
    # Configuración del gráfico
    axes[idx].set_title(f"Tipo: {nombre}", fontsize=14)
    axes[idx].set_xlabel("u", fontsize=12)
    axes[idx].set_ylabel("v", fontsize=12)
    axes[idx].grid(True, linestyle='--', alpha=0.6)
    axes[idx].set_aspect('equal')
    axes[idx].set_xlim(-10, 10)
    axes[idx].set_ylim(-10, 10)

plt.tight_layout()
plt.show()
```
